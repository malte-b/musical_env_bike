## Raincloud Plots for PM columns

The following code is inspired by https://wellcomeopenresearch.org/articles/4-63

```python
import ptitprince as pt
import seaborn as sns
import matplotlib.collections as clt
```

```python
pm1 = pm1.to_frame()
pm1['group'] = "PM1_Values"

pm2_5 = pm2_5.to_frame()
pm2_5['group'] = "PM2.5_Values"

pm10 = pm10.to_frame()
pm10['group'] = "PM10_Values"

pm1.columns = ['value', 'group']
pm2_5.columns = ['value', 'group']
pm10.columns = ['value', 'group']

frames = [pm1, pm2_5, pm10]
df_pm = pd.concat(frames)

df_pm
```

```python
# plotting the clouds
sns.set(style="whitegrid",font_scale=2)
f, ax = plt.subplots(figsize=(20, 10))
dy="group"; dx="value"; ort="h"; pal = "Set1"
ax=pt.half_violinplot( x = dx, y = dy, data = df_pm, palette = pal,
       bw = .2, cut = 0.,scale = "area", width = .6, inner = None,
      orient = ort)

# adding the rain and the jitter
ax=sns.stripplot( x = dx, y = dy, data = df_pm, palette = pal, 
     edgecolor = "white",size = 3, jitter = 1, zorder = 0,
     orient = ort)

#adding the boxplot with quartiles
ax=sns.boxplot( x = dx, y = dy, data = df_pm, color = "black", 
       width = .15, zorder = 10, showcaps = True,
       boxprops = {'facecolor':'none', "zorder":10}, showfliers=True,
       whiskerprops = {'linewidth':2, "zorder":10},
       saturation = 1, orient = ort)
```
