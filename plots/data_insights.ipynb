```python
df_ride_1 = pd.read_csv('../data/ride1.csv')

df_ride_1
```

```python
df_ride_2 = pd.read_csv('../data/ride2.csv')
df_ride_3 = pd.read_csv('../data/ride3.csv')
df_ride_4 = pd.read_csv('../data/ride4.csv')
df_ride_5 = pd.read_csv('../data/ride5.csv')
df_ride_6 = pd.read_csv('../data/ride6.csv')
df_ride_7 = pd.read_csv('../data/ride7.csv')
```

## Plots for each ride

```python
df_ride_1.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 1')
df_ride_2.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 2')
df_ride_3.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 3')
df_ride_4.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 4')
df_ride_5.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 5')
df_ride_6.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 6')
df_ride_7.plot(kind = 'line', y = ['PM1.0', 'PM2.5', 'PM10.0'], title = 'Ride 7')
```

## Pltots for each PM (particulate matter)
